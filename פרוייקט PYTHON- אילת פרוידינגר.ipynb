```python
#Now let's check what are the leading territories:
```

```python
grouped=Merge_Table_2.groupby(['Name','Year'])['SalesOrderID'].nunique().reset_index()
grouped_sorted=grouped.sort_values(by='SalesOrderID', ascending=False)
plt.figure(figsize=(15, 6))
plot=sb.barplot(data=grouped, x='Name',y='SalesOrderID', hue='Year', order=grouped_sorted['Name'].value_counts().index)
for p in plot.patches:
    plot.annotate(format(p.get_height(), '.0f'),
                  (p.get_x() + p.get_width() / 2., p.get_height()),
                  ha='center', va='center',
                  xytext=(0, 5), 
                  textcoords='offset points')
plt.xlabel('Territory', fontsize=12)
plt.ylabel('Distinct Count of Sales Orders', fontsize=12)
plt.title('Distinct Count of Sales Orders by Territory and Year', fontsize=14)
plt.show()
```

```python
#We will check the correation also via graphs
```

```python
distinct_counts=Merge_Table_2.groupby(['Name', 'Year']).agg({
    'SalesOrderID': 'nunique',
    'BusinessEntityID': 'nunique'
}).reset_index()
plt.figure(figsize=(12, 8))
years=distinct_counts['Year'].unique()
for i, year in enumerate(years, start=1):
    plt.subplot(2, 2, i)
    year_data = distinct_counts[distinct_counts['Year'] == year]
    sb.scatterplot(data=year_data, x='SalesOrderID', y='BusinessEntityID', hue='Name', s=100)
    plt.title(f'Year {year}')
    plt.xlabel('Distinct Count of Sales Orders')
    plt.ylabel('Distinct Count of Sales Person')
plt.tight_layout()
plt.show()
```

```python
#And as expected, no corellation is seen in the graphs between sales orders and sales person- the points are scattered randomly on the graphs. 
```

```python
#Let's check what are the leading months per year:
```

```python
grouped2=Merge_Table_2.groupby(['Month','Year'])['SalesOrderID'].nunique().reset_index()
plt.figure(figsize=(15, 6))
plot2=sb.barplot(data=grouped2, x='Month',y='SalesOrderID', hue='Year')
for p in plot2.patches:
    plot2.annotate(format(p.get_height(), '.0f'),
                  (p.get_x() + p.get_width() / 2., p.get_height()),
                  ha='center', va='center',
                  xytext=(0, 5), 
                  textcoords='offset points')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Distinct Count of Sales Orders', fontsize=12)
plt.title('Distinct Count of Sales Orders by Month and Year', fontsize=14)
plt.show()
```

```python
#Based on the graph there is no seasonality- the only years with full data are 2012 and 2013. 
#While the second half of the year is significantly higher in sales orders in 2013, in 2012 I don't see this trend.
#It is interesting to check if this increase in sales during the second half of 2013, has occured in all territories.
```

```python
grouped3=Merge_Table_2.groupby(['Month', 'Year', 'Name'])['SalesOrderID'].nunique().reset_index()
unique_names=grouped3['Name'].unique()
for name in unique_names:
    plt.figure(figsize=(10, 3))
    subset=grouped3[grouped3['Name']==name]
    plot3=sb.barplot(data=subset, x='Month',y='SalesOrderID', hue='Year')
    for p in plot3.patches:
        plot3.annotate(format(p.get_height(), '.0f'),
                  (p.get_x() + p.get_width() / 2., p.get_height()),
                  ha='center', va='center',
                  xytext=(0, 5), 
                  textcoords='offset points')
    plt.xlabel('Month', fontsize=12)
    plt.ylabel('Distinct Count of Sales Orders', fontsize=12)
    plt.title(f'Distinct Count of Sales Orders by Month and Year for "{name}"', fontsize=14)
    plt.legend(title='Year')
    plt.tight_layout()
    plt.show()
```

```python
#This increase in sales orders occures in most territories but not all of them.
#It is intersting to check why it happened but I will not contiunue to analyze it in this project.
```
